**FUNCTION**

A function is a block of code which only runs when it is called.

You can pass data, known as parameters into a function.

A function can return data as a result.

```python
#syntax


def function_name(name,age):

  print(name)
  print(age)


  return name

print (function_name('shalu',20))
```

```python
#function without argument


def function_name():
  a = int(input('enter number1 : '))
  b = int(input('enter number2 : '))


  return a+b
print (function_name())
```

```python
#function with argument

def function_name(c,d):
  c = int(input('enter number1 : '))
  d = int(input('enter number2 : '))

  return c+d

print(function_name(40,20))
```

```python
#function without argument and without return type


def function_name():
  e = int(input('enter the num1 : '))
  f = int(input('enter the num2 : '))
  sum = e+f
  print(sum)


print(function_name())
```

```python
#sub

def sub(a,b):
  if a>b:
    sub = a-b
  else:
    sub = b-a
  return sub
print(sub(1,9))
```

```python
#default arguments and non default arguments
#non default argument cannot be passed after default argument


def sub(a,b=20):

  if a>b:
    sub = a-b
  else:
    sub = b-a
  return sub
print(sub(9))
```

```python
def sub(b,c=45,f):

  if b>c:
    sub = b-c
  else:
    sub = c-b
  return sub
print(sub(5))


#non default argument cannot be passed after default argument
```

```python
#automatic memory management
def add(a,b=20,c=10,d=4):
  add = a+b+c+d
  return add
print(add(2,50,22,4))

```

```python
#factorial

def factorial(num):
  factorial = 1
  if num<0:
    print('factorial does not exist for negative numbers')
  elif num==0:
    print('factorial of zero is 1')
  else:
    for i in range(1,num+1):
      factorial = factorial * i
  return factorial
print(factorial(5))
```
